```python
import torch

from dataset.texts import valid_symbols
from lightspeech import FeedForwardTransformer
from utils.hparams import load_hparam_str
```

```python
ckpt = torch.load("/home/hadware/Code/ogmios-workbench/LightSpeech/checkpoints/ljspeech/ljspeech_fastspeech_d9290f7_40k_steps.pyt", map_location=torch.device('cpu'), weights_only=True)
hp = load_hparam_str(ckpt['hp_str'])
idim = len(valid_symbols)
odim = hp.audio.num_mels
model = FeedForwardTransformer(idim, odim, hp)
```

```python
model.eval()
```

```python
dummy_input = torch.randint(0, len(valid_symbols), size=(50,)).to(dtype=torch.int64)
torch.onnx.export(model, 
                  dummy_input, 
                  "test-export.onnx", 
                  opset_version=12, 
                  do_constant_folding=True, 
                  input_names=["input"], 
                  output_names=["output"])
```
